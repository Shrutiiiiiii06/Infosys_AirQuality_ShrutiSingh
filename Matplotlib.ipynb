```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# line plot
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color = 'purple')
plt.title("Line Plot: Sine Wave")
plt.show()
```

```python
# Scatter Plot
plt.scatter(x,np.cos(x) , color = 'red')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100) # Generate 100 points between 0 and 10
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, color='blue', linestyle='--', marker='o', markersize=4, label='Sine Wave')
plt.plot(x, y2, color='green', linestyle='-', linewidth=2, label='Cosine Wave')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Sine and Cosine Waves")
plt.legend() # Display the legend (Provides a label for the line, 
#useful when plotting multiple lines and creating a legend with plt.legend().)
plt.grid(True) # Add a grid
plt.show()
```

```python
# Bar Chart
categories = ['A', 'B', 'C']
values = [10,20,15]
plt.bar(categories , values , color = 'pink')
plt.title("Bar Chart Example")
plt.show()
```

```python
# Histogram 
data = np.random.randn(1000)
plt.hist(data , bins = 30 , color = 'blue')
plt.title("Histogram Example")
plt.show()
```

```python
# Pie Chart
sizes = [40,35,25]
labels = ['Apple', 'Banana', 'Cherry']
plt.pie(sizes , labels = labels , autopct = '%1.1f%%')
plt.title("Pie Chart Example")
plt.show()
```

```python
# Box Plot
data = np.random.randn(100)
plt.boxplot(data)
plt.title("Box Plot Example")
plt.show()
```

```python
# Step plot 
plt.step(x , np.sin(x) , color = 'brown')
plt.title("Step Plot Example")
plt.show()
```

```python
# Error Bar Plot
x = np.linspace(0,10,100)
y = np.sin(x)
plt.errorbar(x,y,yerr = 0.1,fmt='o')
plt.title("Error Bar Example")
plt.show()
# yerr: The error values for the y-coordinates. This can be a single float (for symmetrical error bars of the same size), 
#an array of floats (for variable symmetrical error bars), or a 2xN array 
#(for asymmetrical error bars, where the first row is lower errors and the second row is upper errors).
# Fmt : format of the data points
#color: The color of the error bars.
#elinewidth: The line width of the error bars.
#capsize: The length of the caps at the ends of the error bars. 
#barsabove: A boolean (True/False) indicating whether error bars should be drawn above or below the data points.
#lolims, uplims, xlolims, xuplims: Booleans to indicate that a value gives only upper/lower limits for the error bars.
```

```python
# Error Bar Plot
x = [1,2,3,4,5]
y = [2,4,1,5,3]
plt.errorbar(x,y,yerr = 0.1,fmt='o')
plt.title("Error Bar Example")
plt.show()
```

```python
# line plot
x = np.linspace(0,10,100)
y = np.sin(x)
plt.fill_between(x,y,color = 'purple')
plt.title("Line Plot: Sine Wave")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate the data (as in your example)
heat = np.random.rand(10, 10)
rows, cols = heat.shape

# 2. Create the heatmap figure
plt.figure(figsize=(8, 8))
plt.imshow(heat, cmap='coolwarm')
plt.colorbar()
plt.title("Heatmap with Values in Boxes")

# 3. Add the values to each box
# Loop over data dimensions and create text annotations.
for i in range(rows):
    for j in range(cols):
        # Format the number to 2 decimal places and center it in the cell
        plt.text(j, i, f'{heat[i, j]:.2f}',
                 ha="center", va="center", color="k") # k is for black color

plt.show()
```
